# WSN Event Statistics by Week

## Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import reporting.services.statistic.event_statistic as event_stat
```

## Load Data

```python
x, y, y2 = event_stat.load_statistic('event_count', event_stat.EVENT_STATISTIC_TYPE_WEEK)
y = [num/1000 for num in y]
```

## Plot Size vs Count

```python
fig=plt.figure(figsize=(10,8))
subplt1=fig.add_subplot(1,1,1)
subplt1.set_title('WSN Event Statistics by Week', fontsize=20)
subplt1.set_xlabel('Week (0-Sunday)')
subplt1.set_ylabel('Size')
subplt1.set_ylim(2000,10000)
for a,b in zip(x,y):
    subplt1.text(a, b+1, '%.0f' % b, ha='center', va='bottom', fontsize=10)
subplt1.bar(x, y, color='#FFD700')

subplt2=subplt1.twinx()
subplt2.set_ylabel('Count')
subplt2.set_ylim(3000,10000)
for a,b in zip(x,y2):
    subplt2.text(a, b+1, '%.0f' % b, ha='center', va='bottom', fontsize=10)
subplt2.plot(x, y2, 'r-.', marker='o')
plt.grid(alpha=0.3)

# plt.legend()
# plt.show()
```
