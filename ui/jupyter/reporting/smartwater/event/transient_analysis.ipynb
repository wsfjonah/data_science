# Transient Analysis

## Setup

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import json
import datetime
```

```python
with open("/var/dws/smartwater/transient/1527034420000-wsntransient-5a6eb4e70b272a1f64fa26b2-pressure.json",'r') as load_f:
    transient = json.load(load_f)
    # print(transient)

# print(transient['tsda']['data'])

timelist = []
valuelist = []

for k, v in transient['tsda']['data'].items():
    timelist.append(datetime.datetime.fromtimestamp(int(k)/1000))
    valuelist.append(v)

fig=plt.figure(figsize=(15,5))

def add_ax(fig, color='black'):
    # ax = fig.add_axes([0.3, 0.2, 0.5, 0.6])
    ax = fig.add_axes([0.385, 0.2, 0.25, 0.7]) # x, y, width, height
    ax.set_axis_off()
    ax.text(0.1, 0.8, 'Hammer\n(78.4, 19.2m, 60s)', color=color,size=16,
            ha='left', va='center', alpha=0.6, transform=ax.transAxes)
    ax.bar([1], [1], width=0.5, facecolor = 'blueviolet', alpha=0.3)
    


subplt1=fig.add_subplot(1,1,1)
subplt1.set_title('Transient Analysis', fontsize=20)
subplt1.set_xlabel('Date Time')
subplt1.set_ylabel('Pressure')
fig.autofmt_xdate()

subplt1.plot(timelist, valuelist)

add_ax(fig)
```
