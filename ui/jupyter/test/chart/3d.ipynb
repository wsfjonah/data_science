```python
import re
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
X = np.array([1,2,3,4,5])
Y = np.array([1,2,3,4,5])

X, Y = np.meshgrid(X, Y)
Z = np.zeros_like(X)
print(X)
print(Y)

Z[1,2]=20
Z[1,3]=30
Z[2,3]=10
Z[4,1]=5
Z[3,2]=6


print(Z)
fig=plt.figure()
ax = fig.add_subplot(111, projection='3d') 
ax.plot_surface(X, Y, Z)
```

```python
X = [1,2,3]  # coordinates of each bar
Y = [2,4,0]  # coordinates of each bar
Z = [0,0,0]  # coordinates of each bar
dx = [0.5, 0.5, 0.5]  #width of each bar
dy = [0.5, 0.5, 0.5]  #depth of each bar
dz = [5, 4, 7]        #height of each bar

fig=plt.figure(figsize=(20,20))
ax = fig.add_subplot(221, projection='3d')
# ax.view_init(elev=0,azim=0)#,,azimzelevy
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.bar3d(X, Y, Z, dx, dy, dz)

ax = fig.add_subplot(222, projection='3d')
ax.view_init(elev=10,azim=0)#,,azimzelevy
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.bar3d(X, Y, Z, dx, dy, dz)
```
